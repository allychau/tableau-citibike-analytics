```python
#glob.glob(pathname, *, recursive=False) Returns a list of pathnames
frames = [ pd.read_csv(f) for f in glob.glob('data/*.csv') ]
```

```python
citibike_clean = pd.concat(frames)
```

```python

citibike_clean.head()
```

```python
# Sort by startime and keep it in place.  keep it as a dataframe instead of panda series
citibike_clean.sort_values("starttime",  ascending=True,inplace=False)
```

```python
citibike_clean.count()
```

```python
# Export the clean citibike trip data into a csv
citibike_clean.to_csv("data/citibike_trip2019.csv",encoding='utf-8', index=False)
```
